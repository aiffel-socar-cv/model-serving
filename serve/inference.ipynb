```python
import os
import requests
import json
import base64
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_DIR = "/Users/tseo/Documents/Github/viai-serving/DATA/originals"
```

```python
file_list = []
for root, dirs, files in os.walk(DATA_DIR):
    for file in files:
        file_list.append(os.path.join(root, file))

file_list[:5]
```

```python
data_types = ['dent', 'scratch', 'spacing']

df_dent = pd.DataFrame(columns=['class', 'mask', 'confidence'])
df_scratch = pd.DataFrame(columns=['class', 'mask', 'confidence'])
df_spacing = pd.DataFrame(columns=['class', 'mask', 'confidence'])

def save_dataframe(result, data_type):
    global df_dent
    global df_scratch
    global df_spacing

    data_str = result.decode('utf8')
    data_dict = eval(data_str)

    if data_type == "dent":
        df_dent = df_dent.append(data_dict, ignore_index=True)
    if data_type == "scratch":
        df_scratch = df_scratch.append(data_dict, ignore_index=True)
    if data_type == "spacing":
        df_spacing = df_spacing.append(data_dict, ignore_index=True)

def inference_data(file_list, data_type):
    print(f"{data_type} inferring...")
    tot_len = len(file_list)

    for idx, image in enumerate(file_list):
        image_name = image.split('/')[-1] # file name
        with open(image, 'rb') as img:
            base64_string = base64.b64encode(img.read())
        data = {'name': image_name, 'body': base64_string} # data to be posted
        res = requests.post(f"http://0.0.0.0:9000/predictions/{data_type}", data=data)

        if res.status_code == 200:
            print(f"[{idx + 1}/{tot_len}] {image_name} inffered done.")
        else:
            print(f"{image_name} inffered failed.")
            break

        result = res.content
        save_dataframe(result, data_type)


```

```python
for data_type in data_types:
    inference_data(file_list, data_type)
```
